```python
def findrho(p,r,vo_a,rc_r0):
    f1=mp.log(np.log(p)+(0.5*vo_a**2/p**2/r**4 ))
    f2=np.log( -2.*rc_r0*(1-1/r) )
    return f1-f2

radius=np.arange(1,100,0.05) #normalised to r0
rho=np.zeros((len(radius))) #normalised to base density


rc_r0=2.0
vo_a=(rc_r0)**2*np.exp(-2*rc_r0+3./2)
#r_critical=const.G.value*mass_sun/2./v_critical_sq/r0 #normalised to solar radius

guess=1.
for count,ii in enumerate(radius):
        
    root=optimize.root(findrho,guess,args=(ii,vo_a,rc_r0),tol=1e-40)
    guess=root.x[0]
    rho[count]=root.x[0]

hydro_stat=np.exp(-2*rc_r0*(1-1/radius))
```

# Question 3

```python
# Mathematics from: New views of the solar wind with the Lambert W function, S. Cranmer, 2004
# https://ui.adsabs.harvard.edu/abs/2004AmJPh..72.1397C/abstract

def findw(w,r,r_critical):
    Dr=(r/r_critical)**(-4)*np.exp(4*(1-r_critical/r)-1)
    return w*np.exp(w)+Dr


mass_sun=const.M_sun.value
radius_sun=const.R_sun.value
factor=const.k_B.value/const.m_p.value


radius=np.arange(1.0,200,0.5)
temp=np.array((0.1,0.2,0.5,1.,2.,5.))*1e6 #kelvin
velocity=np.zeros((len(radius),6))
v_critical_sq=(factor*temp*2.)
r_critical=const.G.value*mass_sun/2./v_critical_sq/radius_sun #normalised to solar radius

for tcount,jj in enumerate(temp):
   
    for count,ii in enumerate(radius):
        if ii <= r_critical[tcount]:
            guess=2.
        else:
            guess=-2.
        root=optimize.fsolve(findw,guess,args=(ii,r_critical[tcount]))
        guess=root[0]
        velocity[count,tcount]=root[0]
```

```python
plt.plot(radius,np.sqrt(-v_critical_sq*velocity)/1e3)
plt.plot(r_critical,v_critical_sq**0.5/1e3,'o')
plt.xscale('log')
plt.xlabel('r/R$_\odot$',fontsize=18)
plt.xticks(fontsize=16)
plt.yscale('log')
plt.ylabel('v (km/s)',fontsize=18)
plt.yticks(fontsize=16)
plt.ylim(1.2,1.1*10**3)
```

```python
def findwz(w,x):
    return w*np.exp(w)-x

x=np.arange(-1/np.exp(1),-1e-2,0.0035)
x2=np.arange(-1/np.exp(1),6,0.006)
tlen=len(x)+len(x2)

w=np.zeros(tlen)
for count,ii in enumerate(x):
        guess=-2.
        root=optimize.fsolve(findwz,guess,args=(ii))
        guess=root[0]
        w[count]=root[0]

for count,ii in enumerate(x2):
        guess=2.
        root=optimize.fsolve(findwz,guess,args=(ii))
        guess=root[0]
        w[count+len(x)]=root[0]
```

```python
plt.plot(np.concatenate((x,x2)),w,'.')
plt.plot([0,0],[-7,2],'--',alpha=0.5,color='black')
plt.plot([-1,6],[0,0],'--',alpha=0.5,color='black')
plt.locator_params(axis='y', nbins=10)
plt.xlabel('z',fontsize=18)
plt.xticks(fontsize=16)
plt.ylabel('W(z)',fontsize=18)
plt.yticks(fontsize=16)
```
